```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification,make_blobs,make_moons
from sklearn.model_selection import train_test_split
```

```python
x1,y1=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_informative=2,n_clusters_per_class=1,class_sep=2.0,random_state=42)
```

```python
plt.scatter(x1[:,0],x1[:,1],c=y1)
plt.show()
```

```python
x2,y2=make_blobs(n_samples=1000,n_features=2,centers=2,random_state=45)
```

```python
plt.scatter(x2[:,0],x2[:,1],c=y2)
plt.show()
```

```python
x3,y3=make_moons(n_samples=1000,noise=0.3,random_state=45)
```

```python
plt.scatter(x3[:,0],x3[:,1],c=y3)
plt.show()
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.3,random_state=56)
```

```python
from sklearn.svm import LinearSVC
model=LinearSVC(random_state=0,tol=1e-5)
model.fit(x1_train,y1_train)
```

```python
y_pred=model.predict(x1_test)
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,accuracy_score
```

```python
confusion_matrix(y1_test,y_pred)
```

```python
precision_score(y1_test,y_pred)
```

```python
recall_score(y1_test,y_pred)
```

```python
accuracy_score(y1_test,y_pred)
```

```python
model.score(x1_train,y1_train)
```

```python
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=0.3,random_state=56)
```

```python
model2=LinearSVC(random_state=0,tol=1e-5)
model2.fit(x2_train,y2_train)
```

```python
y_pred2=model2.predict(x2_test)
```

```python
accuracy_score(y2_test,y_pred2)
```

```python
model2.score(x2_train,y2_train)
```
